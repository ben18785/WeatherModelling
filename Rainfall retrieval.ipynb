Trying to extract rainfall data from Bobo, Burkina Faso. ei here means interim dataset; param 228 is for total precipitation; 205 is runoff (surface and sub); looks like surface runoff is available only until 2010 in the interim daily/LAND dataset; 
stream 'oper' is for the Atmospheric model; type 'an' is for analysis; lat and lon is given as N/W/S/E indicating a grid.

Note that there are also data available for surface runoff which is the depth of water that would accumulate if rainfall was distributed evenly over an area

Description of the surface runoff calculation
https://www.ecmwf.int/sites/default/files/elibrary/2016/17117-part-iv-physical-processes.pdf#subsection.H.6.3

Provides a description of each variable
http://apps.ecmwf.int/codes/grib/param-db/

There are also a range of soil characteristics. For example, the clay fraction.

Tried steps at 3/6/9/12 to see if I can get it to download.

Shows how to calculate total precipitation: https://confluence.ecmwf.int/display/CKB/ERA-Interim%3A+How+to+calculate+daily+total+precipitation

```python
#!/usr/bin/env python
from ecmwfapi import ECMWFDataServer

server = ECMWFDataServer()
    
server.retrieve({
    'stream'    : "oper",
    'levtype'   : "sfc",
    'param'     : "228.128",
    'dataset'   : "interim",
    'step'      : "12",
    'grid'      : "0.75/0.75",
    'time'      : "00:00:00/12:00:00",
    'date'      : "2014-07-01/to/2014-07-31",
    'type'      : "fc",
    'class'     : "ei",
    'area'      : "12/-5.3/10/-3.3",
    'format'    : "netcdf",
    'target'    : "test-astmh.nc"
    
})
```
